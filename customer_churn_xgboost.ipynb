```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from utils.helpers import encode_data, f1_score
import pandas as pd 
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

df = encode_data(pd.read_csv("data/WA_Fn-UseC_-Telco-Customer-Churn.csv"))  
pd.set_option('display.max_columns', None)
```

```python
features = df.drop(['Churn', 'customerID'], axis=1)
target = df['Churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=.2, random_state=42)
```

```python
xgb = XGBClassifier(n_estimators = 5000, learning_rate = 0.01, verbosity = 1, random_state = 42, early_stopping_rounds=1000)
xgb.fit(x_train, y_train, eval_set=[(x_test, y_test)])
```

```python
y_pred_test = xgb.predict(x_test)
y_pred_train = xgb.predict(x_train)
print(f"Train F1 Score: {f1_score(y_train, y_pred_train)}")
print(f"Test F1 Score: {f1_score(y_test, y_pred_test)}")
```

```python
accuracy = accuracy_score(y_test, y_pred_test)
print(f"Test Accuracy: {accuracy}")
train_accuracy = accuracy_score(y_train, y_pred_train)
print(f"Train Accuracy: {train_accuracy}")
```

Find best hyperparams for XGB

```python
xgb_hypeparams = {
    'n_estimators': [100, 500, 1000],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}

grid_search = GridSearchCV(estimator=XGBClassifier(verbosity=1, random_state=42), param_grid=xgb_hypeparams, cv=3, n_jobs=-1, verbose=2)
grid_search.fit(x_train, y_train)
print(grid_search.best_params_)
```

```python
xgb = XGBClassifier(n_estimators = 100, max_depth = 3, learning_rate = 0.1, verbosity = 0, random_state = 42, early_stopping_rounds=50)
xgb.fit(x_train, y_train, eval_set=[(x_test, y_test)])
```

```python
y_pred_test = xgb.predict(x_test)
y_pred_train = xgb.predict(x_train)
print(f"Train F1 Score: {f1_score(y_train, y_pred_train)}")
print(f"Test F1 Score: {f1_score(y_test, y_pred_test)}")
```

```python
accuracy = accuracy_score(y_test, y_pred_test)
print(f"Test Accuracy: {accuracy}")
train_accuracy = accuracy_score(y_train, y_pred_train)
print(f"Train Accuracy: {train_accuracy}")
```
